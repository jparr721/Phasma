```python
import numpy as np

# Eigen is column major
ROWS = 0
COLS = 1
FEATURES = 2
ENTRIES = 3

# Numpy is row major
NP_ROWS = 2
NP_COLS = 3
NP_FEATURES = 1
NP_ENTRIES = 0

shape = list(map(int, open("input").readline().rstrip().split(' ')))
rows = shape[ROWS]
cols = shape[COLS]
features = shape[FEATURES]
entries = shape[ENTRIES]

shape[NP_ROWS] = rows
shape[NP_COLS] = cols
shape[NP_FEATURES] = features
shape[NP_ENTRIES] = entries

shape
```

```python
def read_file(file_name):
    lines = open(file_name).readlines()[1:]
    for i, line in enumerate(lines):
        lines[i] = list(map(float, line.rstrip().split()))

    np.set_printoptions(linewidth=550, threshold=2000, precision=2)
    lines = np.array(lines)

    entries = []
    for i in range(0, len(lines.T), 6):
        entries.append(lines[:, i:i+3])
        entries.append(lines[:, i+3:i+6])

    lines = np.array(entries)
    lines = lines.reshape(shape)
    return lines

lines = read_file("input")
```
